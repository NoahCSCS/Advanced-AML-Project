```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fj.csv", index_col = "Unnamed: 0")
df["title"]
```

```python
df = df.apply(lambda x: x.apply(lambda y: "" if y=="MISSING" else y))
```

```python
text_col = ["title", "company_profile", "description", "requirements", "benefits"]
df["full_listing"] = df.apply(lambda row: "".join([entry + "\n" for entry in row[text_col]]), axis = 1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words={'english'})
X = vectorizer.fit_transform(df["full_listing"])
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
X @ np.zeros(104164)
```

```python
spacy_scores = [nlp(word)[0].vector for word in vectorizer.vocabulary_.keys()]
```

```python
spacy_scores = np.matrix(spacy_scores)
```

```python
document_scores = spacy_scores @ X
```
